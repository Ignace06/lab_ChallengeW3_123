```python
# Cleanup Name that contain "Mega"

def clean(x):
    if 'Mega' in x:
        text2 = x.split("Mega", 1)
        x1 = text2[1]
        return "Mega" + x1
    else:
        return x

# pokemon['Name'] = list(map(clean, pokemon['Name']))
pokemon['Name'] = pokemon['Name'].apply(clean)
pokemon.head(20)
```

```python
# Identify the pokemon with the highest A/D Ratio.
pokemon['ratio A/D']=pokemon['Attack']/pokemon['Sp. Def']
pokemon.head(10)
```

```python
# Identify the pokemon with the highest A/D Ratio.
pokemon[pokemon['ratio A/D']==pokemon['ratio A/D'].max()]
# DeoxysAttack Forme
```

```python
# Identify the pokemon with the lowest A/D Ratio.
pokemon[pokemon['ratio A/D']==pokemon['ratio A/D'].min()]
# Shuckle
```

```python
# Create a new column called Combo Type whose value combines Type 1 and Type 2.

pokemon['combine']= pokemon['Type 1'].astype(str) + ' ' + pokemon['Type 2'].fillna('') 
pokemon.head(20)    
```

```python
# Identify the pokemon whose A/D Ratio are among the top 5.

pokemon.pivot_table(index=['Name'], values=['ratio A/D']).sort_values(by="ratio A/D",ascending=False).head(5)
```
