We break down the main problem into several sub problems:

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# sub problem 1
# How to aggregate the amount_spent for unique customers?
# orders.pivot_table(index =['CustomerID'], values = ['amount_spent'])
orders.groupby(['CustomerID']).agg({'amount_spent':max}).rank()
```

```python
orders[orders['amount_spent'] == orders['amount_spent'].max()]
```

```python
# How to select customers whose aggregated amount_spent is in a given quantile range
labels = ["VIP", "Preferred", "Moderate", "Low"]
qbins = pd.qcut(orders['amount_spent'],4, labels=labels)
orders['Label'] = pd.qcut(orders['amount_spent'], 4, labels = labels)
qbins.value_counts()
orders
```

```python

# Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?
```

```python
# your code here
orders[orders['Label'] == 'VIP'].groupby(['Country']).agg({'CustomerID':pd.Series.nunique}).sort_values(by= 'CustomerID', ascending=False).head(5)
# UK has the most VIP customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
orders[orders['Label'].isin(['VIP','Preferred'])].groupby(['Country']).agg({'CustomerID':pd.Series.nunique}).sort_values(by= 'CustomerID', ascending=False).head(5)
# UK has the most VIP+Preferred customers
```
